```python
words = ['someone','homesickness','coolness','legality','reading','addition','friday','firework','whiteness','joviality']

neutral = ['someone','legality','reading','addition','whiteness']         
emotional = ['homesickness','friday','firework','joviality','coolness']

# list of all unique characters in all columns
```

```python
# read google forms responses in
forms_resps = pd.read_csv('/projects/MINDLAB2022_MR-semantics-of-depression/scripts/bachelor_scripts/analysis/word_stim_tsv_files/google_forms_resps_19.csv', sep=';')
```

```python
forms_resps
```

```python
# compute mean for all columns with names not endning with .1
valence_means = forms_resps.filter(regex='^(?!.*\.1$).*$').mean()
valence_means
```

```python
# compute mean for all columns with names ending with .1
arousal_means = forms_resps.filter(regex='.*\.1$').mean()
arousal_means
```

```python
column_names = ['Someone','Homesickness.2','Coolness.2','Legality.2','Reading.2','Addition.2','Friday.2','Whiteness.2','Firework.2','Joviality.2']
column_names

for i in column_names:
    print(forms_resps[i].value_counts())
    
```

```python
# count number of different unique responses for the last 10 columns in the dataframe and percentages
for i in column_names:
    print(forms_resps[i].value_counts(normalize=True))
```

```python
# make valence_means a list
valence = valence_means.tolist()
arousal = arousal_means.tolist()
words = words 
```

```python
# combine valence, arousal, and words to dataframe
df = pd.DataFrame(list(zip(words, valence, arousal)), columns =['Words', 'Valence', 'Arousal'])
df
```

```python
df['trial_type'] = 'Neutral'
df.loc[(df['Arousal'] > 2) & (df['Valence'] < 3), 'trial_type'] = 'Emotional' 
df.loc[(df['Arousal'] > 2) & (df['Valence'] > 6), 'trial_type'] = 'Emotional'
```

```python
df # in the binary response, people disagree with reading (80N/ 20E), firework (60E/40E), whiteness (30N, 70E) 

# rename "Words" column to "word"
df.rename(columns={'Words':'Word'}, inplace=True)
df.rename(columns={'Valence':'V.Mean.Sum'}, inplace=True)
df.rename(columns={'Arousal':'A.Mean.Sum'}, inplace=True)
```

```python
# add Word column from df to xanew as rows
xanew = xanew.append(df, ignore_index=True)
```
